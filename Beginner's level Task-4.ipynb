## Gray Scale Image

```python
img_gray=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY) # Normal Image
plt.figure(figsize=(5,5))
plt.imshow(img_gray,cmap="gray")
plt.title("GrayScale Image")
plt.show()

img_gray1=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY) # Digital Image
plt.figure(figsize=(5,5))
plt.imshow(img_gray1,cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

## Inverted  Image

```python
img_invert=cv2.bitwise_not(img_gray) # Normal Image
plt.figure(figsize=(5,5))
plt.imshow(img_invert,cmap="gray")
plt.title("inverted image")
plt.axis("off")
plt.show()

img_invert1=cv2.bitwise_not(img_gray1) # digital Image
plt.figure(figsize=(5,5))
plt.imshow(img_invert1,cmap="gray")
plt.title("inverted image")
plt.axis("off")
plt.show()
```

## Blur Image

```python
img_smoothing=cv2.GaussianBlur(img_invert,(23,23),sigmaX=0,sigmaY=0) # Normal Image
plt.figure(figsize=(5,5)) 
plt.imshow(img_smoothing,cmap="gray")
plt.title("Blur Image")
plt.axis("off")
plt.show()


img_smoothing1=cv2.GaussianBlur(img_invert1,(23,23),sigmaX=0,sigmaY=0) # Digital Image
plt.figure(figsize=(7,7))
plt.imshow(img_smoothing1,cmap="gray")
plt.title("Blur Image")
plt.axis("off")
plt.show()
```

## Pencil Sketch

```python
final_img=cv2.divide(img_gray,255-img_smoothing,scale=255) # Normal Image
plt.figure(figsize=(5,5))
plt.imshow(final_img,cmap="gray")
plt.title("Final sketch image")
plt.axis("off")
plt.show()

final_img1=cv2.divide(img_gray1,255-img_smoothing1,scale=255) # Digital Image
plt.figure(figsize=(5,5))
plt.imshow(final_img1,cmap="gray")
plt.title("Final sketch image")
plt.axis("off")
plt.show()


```
