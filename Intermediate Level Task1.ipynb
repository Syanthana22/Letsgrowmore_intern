```python
def plot(area,name):
    middleEastTerrorism=Terrorism[Terrorism['region']==area]
    middleEastTerrorism.iyear.plot(kind='hist',bins=30,figsize=(10,5),color='red',fontsize=15)
    plt.xlabel('year',fontsize=15)
    plt.ylabel('Freuquency',fontsize=15)
    plt.title(name)
    plt.show()
```

```python
for num in dic.keys():
    plot(num,dic[num])
```
